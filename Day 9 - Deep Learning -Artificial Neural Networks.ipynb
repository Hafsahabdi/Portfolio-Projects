# Classification

```python
# x=df[['Pregnancies', 'Glucose', 'BloodPressure','SkinThickness','BMI']]
```

```python
x=df.iloc[:,0:8]
y=df.iloc[:,8]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model=Sequential()
model.add(Dense(12,activation='relu')) #input layer
model.add(Dense(8,activation='relu')) #Hidden layers
model.add(Dense(1,activation='sigmoid'))#  Output layer 
#Binary classification 1/0 yes or no, up down, right or left sigmoid otherwise softmax
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(x,y,epochs=150,batch_size=10,verbose=1)
```

```python
model.summary()
```

```python
scores=model.evaluate(x,y)
```

```python
scores[0]
```

```python
"Accuracy:   " +str(scores[1])
```

```python
# Deep Neural Network
```

```python
model=Sequential()
model.add(Dense(60,activation='relu')) #input layer
model.add(Dense(80,activation='relu')) #Hidden layers
model.add(Dense(30,activation='relu'))
model.add(Dense(16,activation='relu'))
model.add(Dense(8,activation='relu'))
model.add(Dense(1,activation='sigmoid'))#  Output layer 
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(x,y,epochs=100,batch_size=32,verbose=1)
```

```python
model.summary()
```

```python
history=model.fit(x,y,epochs=120,batch_size=32,verbose=1)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(history.history['accuracy'])
```

```python
# optimum early stopping
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint
```

```python
filepath="weights.best.hdf5"
checkpoint=ModelCheckpoint(filepath,monitor='accuracy',verbose=1,save_best_only=True,mode="max")
callback_list=[checkpoint]
```

```python
model.fit(x,y,validation_split=0.20,epochs=150,batch_size=32,callbacks=callback_list,verbose=0)
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stopping = EarlyStopping()
```

```python
model.fit(x,y,validation_split=0.20,epochs=150,batch_size=32,callbacks=early_stopping,verbose=1)
```

```python
scores=model.evaluate(x,y)
```

# Regression

```python
import pandas as pd 
```

```python
df=pd.read_pickle('kc_house.pkl')
```

```python
df.head()
```

```python
x=df.drop('price',axis=1)
y=df['price']
```

```python
# normalizing the data
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
columns=x.columns
```

```python
x=scaler.fit_transform(x)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model=Sequential()
model.add(Dense(81,activation='relu')) #input layer
model.add(Dense(40,activation='relu')) #Hidden layers
model.add(Dense(27,activation='relu')) #Hidden layers
model.add(Dense(19,activation='relu')) #Hidden layers
model.add(Dense(5,activation='relu')) #Hidden layers
model.add(Dense(1))#  Output layer 
model.compile(loss='mse', optimizer='adam')
```

```python
model.fit(x_train,y_train,validation_data=(x_test,y_test),batch_size=128,epochs=100,verbose=1)
```

```python
model.summary()
```

```python
pred=model.predict(x_test)
```

```python
r2_score(y_test,pred)
```

```python
mean_squared_error(y_test,pred)**0.5
```
