# Data Visualization

```python
# Import Packages
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=np.arange(20) # generates 20 numbers
y=np.random.normal(10,1,20) # generates 20 numbers std=1 and mean is 10
z=np.random.normal(10,2,20) # generates 20 numbers std=2 and mean is 20
k=np.random.normal(10,3,20) # generates 20 numbers std=3 and mean is 20
```

```python
plt.plot(x,y,color='b', label='y');
plt.plot(x,z,color='r', label='z');
plt.plot(x,k,color='g', label='k');
plt.legend()
```

```python
fig = plt.figure(figsize=(20,5))
ax = fig.add_subplot(131)
ax.set_title('X vs Y')
ax.set_xlabel('x data')
ax.set_ylabel('y values')
ax.plot(x,y)

bx = fig.add_subplot(132)
bx.set_title('X vs Z')
bx.set_xlabel('x data')
bx.set_ylabel('z values')
bx.plot(x,z)


cx = fig.add_subplot(133)
cx.set_title('X vs K')
cx.set_xlabel('x data')
cx.set_ylabel('k values')

cx.plot(x,k);
```

```python
fig=plt.figure(figsize=(20,10))
ax=fig.add_subplot(231)
ax.set_title('X vs Y')
ax.set_xlabel('X values')
ax.set_ylabel('Y values')
ax.plot(x,y)

bx=fig.add_subplot(232)
bx.set_title('X vs Z')
bx.set_xlabel('X values')
bx.set_ylabel('Z values')
bx.plot(x,z)

cx=fig.add_subplot(233)
cx.set_title('X vs K')
cx.set_xlabel('X values')
cx.set_ylabel('K values')
cx.scatter(z,k);


ex=fig.add_subplot(234)
ex.set_title('X vs K')
ex.set_xlabel('X values')
ex.set_ylabel('K values')
ex.plot(x,k);

dx=fig.add_subplot(235)
dx.set_title('X vs K')
dx.set_xlabel('X values')
dx.set_ylabel('K values')
dx.plot(x,k);

fx=fig.add_subplot(236)
fx.set_title('X vs K')
fx.set_xlabel('X values')
fx.set_ylabel('K values')
fx.boxplot([y,z,k]);


```

## feature engineering 
generating new columns 

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
df['day']=(df['Date']).dt.day
df['month']=(df['Date']).dt.month
df['year']=(df['Date']).dt.year
```

```python
df['month_name']=(df['Date']).dt.month_name()
df['weekday']=(df['Date']).dt.day_name()
```

```python
df.head()
```
